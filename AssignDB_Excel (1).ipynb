```python
data['Customer_Gender']= data['Customer_Gender'].fillna(data['Customer_Gender'].mode()[0])
```

```python
data['Country']= data['Country'].fillna(data['Country'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data['Product'].mode()[0]
```

```python
data['State']= data['State'].fillna(data['State'].mode()[0])
data['Product_Category']= data['Product_Category'].fillna(data['Product_Category'].mode()[0])
data['Sub_Category']= data['Sub_Category'].fillna(data['Sub_Category'].mode()[0])
data['Product']= data['Product'].fillna(data['Product'].mode()[0])
```

```python
data.isnull().sum()> 0
```

```python
data['Order_Quantity']= data['Order_Quantity'].fillna(data['Order_Quantity'].mean())
data['Unit_Cost']= data['Unit_Cost'].fillna(data['Unit_Cost'].mean())
data['Unit_Price']= data['Unit_Price'].fillna(data['Unit_Price'].mean())
data['Profit']= data['Profit'].fillna(data['Profit'].mean())
data['Cost']= data['Cost'].fillna(data['Cost'].mean())
data['Revenue']= data['Revenue'].fillna(data['Revenue'].mean())
```

```python
data.isnull().sum()
```

```python
data.to_csv('Sales_process1.csv')
```
